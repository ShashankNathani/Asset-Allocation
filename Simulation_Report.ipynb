We will now test multiple asset allocation strategies on our existing dataset for both light-tailed and heavy-tailed distribution. For testing the strategies we fix a rolling window of ${M=120}$ here. After applying these strategies we test these strategies based on three metrics namely - ${ISR}$, ${OSR}$ and ${Turnover}$. We describe briefly how these metrics are calculated.  

$$
\underline{\textbf{ISR}} \\[0.25in]
ISR \ = \ \frac{\hat{μ}_{ISR} \ - \ r} {\hat{σ}_{ISR}}
$$

where ${\hat{µ}_{ISR}}$ denotes the sample average of the return in-sample µt achieved by the portfolio over all time steps from ${t = M}$ to ${T}$, and ${\hat{σ}_{ISR}}$ denotes the corresponding sample standard deviation. Specifically, in-sample ${µ_t}$ is esimated by applying different strategies using estimation with rolling window technique at step ${t}$, we use the excess returns of all N risky assets in steps ${t − M − 1}$ to ${t}$ to estimate and solve for the optimal weight ${w_t}$. Then we calculate the return at time ${t}$ : <br> <br>

$$
\mu_t \ = \ ∑_{i=1}^{N} w^{(t)}_{asset \ i} \ 𝙭 \ \text{return of asset ${i}$ at time ${t}$ } 
$$

$$
\underline{\textbf{OSR}} \\[0.25in]
OSR \ = \ \frac{\hat{μ}_{OSR} \ - \ r} {\hat{σ}_{OSR}}
$$

where ${\hat{µ}_{OSR}}$ denotes the sample average of the return in-sample µt achieved by the portfolio over all time steps from ${t = M+1}$ to ${T}$, and ${\hat{σ}_{OSR}}$ denotes the corresponding sample standard deviation. Specifically, in-sample ${µ_t}$ is esimated by applying different strategies using estimation with rolling window technique at step ${t}$, we use the excess returns of all N risky assets in steps ${t − M }$ to ${t-1}$ to estimate and solve for the optimal weight ${w_t}$. Then we calculate the return at time ${t}$ : <br> <br>

$$
\mu_t \ = \ ∑_{i=1}^{N} w^{(t)}_{asset \ i} \ 𝙭 \ \text{return of asset ${i}$ at time ${t}$ } 
$$

$$
\underline{\textbf{Turnover}} \\[0.25in]
Turnover \ = \ \frac{\sum_{t=M+1}^{T} \sum_{j=1}^{N} |w^{(t+1)}_j - w̃^{(t+1)}_j| }{T-M}
$$

where ${w̃^{(t)}}$ , ${w^{(t)}}$ denotes the weight vector before and after rebalancing at time ${t}$ respectively. <br><br><br>

$$
\underline{\textbf{Asset Allocation Methodologies}} \\[0.3in]
\text{We now go over different asset allocation methodologies and to see how weights are calulcated for allocation in current context.} \\ \text{We fix the rolling window parameter ${M=120}$ for now. For each rolling window period we calculate ${\hatμ}$ and ${\hatΣ}$ as the covariance matrix from the ${120}$ period data. We will have our excess returns as ${\hatμ' = (\hatμ - re)}$.} \\ \text{We incorporate ${\kappa = 0.0125}$ as our required rate of return.} \\[1in]
$$

$$
\underline{\textbf{1.Equally Weigthed (ew)}}  \\[0.3in]
\text{We assign equal weights to each asset at the end of each rebalancing period. Weight for each asset is thus, given by :} \\
w_{i,t} \ = \ \frac{1}{N} \\ \text{with N being total number of risky assets and ${t}$ as the end of each rebalancing period.} \\ 
\text{Here we do not allocate to our risk-free asset}\\[1in]
$$

$$
\underline{\textbf{2.Market Portfolio/Factor Portfolio (mkt)}}  \\[0.3in] 
\text{In this allocation, our entire allocation goes into the market portfolio or the factor portfolio. For us it is the factor asset in this case.} \\
\text{Since entire allocation in one single asset no rebalanceing is required and the reutrn of the factor asset will be return of our strategy.} \\ 
w_{i,t} \ = \ \begin{cases}   1 & \text{for } i= factor\\    
  0 & \text{otherwise}  \end{cases} \\[1in]
$$

$$
\underline{\textbf{3.Sample Based Mean Variance (mv)}} \\[0.3in]
\text{Here we minimize the variance of our returns such that our returns are greater than or equal to a certain target ${\kappa}$ } \\
\text{We have the following optimization problem :} \\ 
\max_{w\in \mathbb{R}^d} \ -w^T\hatΣw \\ 
s.t. \quad r \ + \  \hat\mu'^Tw \ \geq \kappa \\ \text{where excess returns } \hat\mu' \ = \ (\hat\mu \ - \ re) \\[0.2in]
\text{We have the following closed form solution for weight vector :} \\
w =  \frac{(\kappa-r)}{\hatμ'^T\hatΣ^{-1}\hatμ'}\hatΣ^{-1}\hatμ' \\
\text{at the end of each rebalancing period .} \\[1in]
$$

$$
\underline{\textbf{4.Sample Based Unbiased Mean Variance (u-mv)}} \\[0.3in]
\text{In the above expression our current estimate of optimal allocation proportion ${\hatΣ^{-1}(\hatμ')}$ is biased} \\
\text{We can plug in unbiased estimator as follows : } \\[0.2in]
\frac{M-N-2}{M-1}\hatΣ^{-1}(\hatμ') \quad \quad N=10 \quad M=120 \\[0.3in]
\text{We get the following optimum allocation : } \\[0.3in] 
w =  (\frac{M-N-2}{M-1})\frac{(\kappa-r)}{\hatμ'^T\hatΣ^{-1}\hatμ'}\hatΣ^{-1}\hatμ' \\[1in]
$$

$$
\underline{\textbf{5.Bayes-Stein (bs)}} \\[0.3in]
\text{James-Stein estimator for the mean returns shrinks data towards a pre-specified point ${μ_0}$ with certain weight.} \\[0.2in]
\hatμ^{JS} = \alphaμ_0 + (1-α)\hatμ \\[0.2in]
\text{This estimate is biased. We seek to add bias so as to acheive a reduction in variance} \\[0.2in]
\text{We chose this point as the one corresponidng to the minimum variance portfolio :} \\[0.2in]
\mu_0 = \frac{\hatμ^T\hatΣ^{-1}e}{e^T\hatΣ^{-1}e}e \\[0.2in] 
\text{We incorporate Bayesian critera to estimate the weight α :} \\[0.2in]
α = \frac{N+2}{N+2+(M-N-2)(\hatμ-μ_0)^T\hatΣ^{-1}(\hatμ-μ_0)} \quad \quad N=10,M=120 \\[0.2in]
\text{Finally, we plug in this estimate in our mean-variance allocation to get the optimum allocation} \\[0.2in] 
w = \frac{(κ-r)}{(\hat\mu^{JS}-re)^T\hatΣ^{-1}(\hat\mu^{JS}-re)}\hatΣ^{-1}(\hat\mu^{JS}-re) \\[1in]
$$

$$
\underline{\textbf{6.Minimum Variance (min)}} \\[0.3in]
\text{In this case we do not impose any constraint on the required rate of return and chose a portfolio that seeks to minimize the variance.} \\ \text{We also impose self-financing condition here otherwise all our allocation would go into the risk free rate.} \\[0.2in] 
\max_{w\in \mathbb{R}^d} \ -w^T\hatΣw \\ 
s.t. \quad e^Tw = 1 \\[0.2in]
\text{This gives us the following allocation as our optimum allocation :} \\[0.2in]
w = \frac{\hat\Sigma^{-1}e}{e^T\hat\Sigma^{-1}e} \\[1in]
\text{Now we consider some of the above strategies with portfolio constraints involved with no short-selling allowed i.e.} \\ \text{our weights cannot be negative.} \\[0.3in]
$$

$$
\underline{\textbf{7.Sample-based mean-variance with no-shortsale (mv-c)}} \\[0.3in]
\text{In this case we have the following optimization probelm} \\[0.2in]
\max_{w\in \mathbb{R}^d} \ -w^T\hatΣw \\ $$
$$
s.t. \quad r \ + \  \hat\mu'^Tw \ \geq \kappa \tag{1}$$
$$ w \geq 0 \tag{2} $$ $$ 1-e^Tw \geq 0 \tag{3} \\  $$
$$ \text{Constraint (2) is to restrict short-sale of risky assets and constraint (3) restricts shorting of risk free rate.} \\[1in] $$ 



$$
\underline{\textbf{8.Bayes-Stein with no-shortsale  (bs-c)}} \\[0.3in]
\text{Here, we solve the following optimization problem :} \\[0.2in]
\max_{w\in \mathbb{R}^d} \ -w^T\hatΣw \\ $$
$$
s.t. \quad r \ + \  {(\hat\mu^{JS}-re)}^Tw \ \geq \kappa \tag{1}$$
$$ w \geq 0 \tag{2} $$ $$ 1-e^Tw \geq 0 \tag{3} \\  $$
$$ \text{We have the same definitions as the above Bayes-Stein estimator: }\\[0.2in] 
\hatμ^{JS} = \alphaμ_0 + (1-α)\hatμ \\[0.2in]
\mu_0 = \frac{\hatμ^T\hatΣ^{-1}e}{e^T\hatΣ^{-1}e}e \\[0.3in] 
α = \frac{N+2}{N+2+(M-N-2)(\hatμ-μ_0)^T\hatΣ^{-1}(\hatμ-μ_0)}  \\[1in]
$$ 


$$
\underline{\textbf{9.Minimum-variance with no-shortsale (min-c)}} \\[0.3in]
\text{Here, we solve the following :} \\
\max_{w\in \mathbb{R}^d} \ -w^T\hatΣw \\ 
s.t. \quad e^Tw = 1 \\[0.2in]
w \geq 0 \\[1in]
$$

$$
\underline{\textbf{Robust Portfolios}} \\[0.3in]
\text{In these models we have multiople prior. The multi-prior means that instead of specifying a single distribution ${\mathbb P}$ there is a set of } \\ \text{ distributions ${\mathcal{P}}$ called uncertainity set. The decision maker selects the worst distribution and solves max-min problem :} \\ 
\max_{w\in \mathbb{R}} \min_{\mathbb{P}\in\mathcal{P}} -Var^{\mathbb{P}} (w^TR) \\
s.t. \min_{\mathbb{P} \in \mathcal{P}} \mathbb{E}^{\mathbb{P}}[w^TR] \geq κ \\[0.2in] 
\text{Two common ways to specify the uncertainty set are parameter uncertainty and distributional uncertainty} \\[0.2in]
\text{Parameter Uncertainity:} \quad \mathcal{P} = \text{{ mean μ and covariance Σ, (μ,Σ) ∈ 𝑈 ⊂ ${ℝ^d}$ x ${𝕊^d_++}$ }} \\[0.2in]
\text{Distibutional Uncertainity:} \quad \mathcal{P} = \{D(ℙ,\hatℙ)\}\leq δ \\
\text{Specification of δ is used to determine the uncertainity set. ${D}$(.,.) measures distance between two distributions.} \\
\text{Method 10 and 11 look at parameter uncertainity while method 12 looks at distributional uncertainity.}
\\[0.5in]
$$


$$
\underline{\textbf{10. Uncertainty in mean with box uncertainty set  (r-m-1)}} \\[0.3in]
\text{In this case we assume that there in uncertaininty related to μ since it is difficult to estimate. We model uncertainity set as :} \\[0.2in]
U = ∏_{k=1}^d [\hat\mu_k - δ_k,\hatμ_k + δ_k] \text{ x } \{Σ\} \\[0.2in] 
\text{In our contex we can model this as following optimization problem :} \\[0.2in]
\max_{w \in ℝ} -w^T\hatΣw \\
s.t. \quad r \ + \ \hatμ'w - δ^T|w| \geq  κ \\
\text{In this case we use δ=0.004} \\[1in]
$$

$$
\underline{\textbf{11. Uncertainty in mean with ellipsoid uncertainty set  (r-m-1)}} \\[0.3in]
\text{In this case we assume that there in uncertaininty related to μ similar to the previous case. We model uncertainity set as :} \\[0.2in]
U = \{\mu:(\mu-\hat\mu)^TΣ^{-1}(\mu-\hat\mu) \leq δ^2 \} \text{ x } \{Σ\} \\[0.2in] 
\text{In our contex we can model this as following optimization problem :} \\[0.2in]
\max_{w \in ℝ} -w^T\hatΣw \\
s.t. \quad r \ + \ \hatμ'w - δ^T\sqrt{w^T\hatΣw} \geq  κ \\
δ=0.004 \\[1in]
$$

$$
\underline{\textbf{12. Distributional robust (dr-mv)}} \\[0.3in]
\text{In this case we assume distributional uncertainity. We solve the following optimization problem :} \\[0.2in]
\min_{w \in ℝ} (\sqrt{w^T\hatΣw} + δ||w||_p)^2 \\ 
s.t. \quad r \ + \hat\mu'w - δ||w||_p \geq κ \\[0.2in]
\text{with p=1,2 ${\quad}$ We solve using p=2} \\[1in]
$$
